Import required libraries

```python
import pandas as pd
```

Read data and rename headers

```python
df = pd.read_csv('data/FIFA_World_Cup.csv', header=None, names=["league", "season", "team", "players_used", "age", "possession", 
    "matches_played", "starts", "minutes", "90s_played", 
    "goals", "assists", "goals_plus_assists", "goals_minus_penalty", 
    "penalty_goals", "penalty_attempts", "yellow_cards", "red_cards", 
    "expected_goals", "non_penalty_xG", "expected_assists", 
    "non_penalty_xG_plus_xA", "progressive_carries", "progressive_passes", 
    "goals_per_90", "assists_per_90", "goals_plus_assists_per_90", 
    "goals_minus_penalty_per_90", "goals_plus_assists_minus_penalty_per_90", 
    "xG_per_90", "xA_per_90", "xG_plus_xA_per_90", "non_penalty_xG_per_90", 
    "non_penalty_xG_plus_xA_per_90", "url"])
```

```python
df
```

Remove the first and second row as both are the header names

```python
df = df.copy()
df = df.iloc[3:]
```

Dropping columns based on observation

```python
df.drop(columns=['league', 'assists','goals_plus_assists','expected_assists','non_penalty_xG_plus_xA','assists_per_90','goals_plus_assists_per_90',
                 'xA_per_90','xG_plus_xA_per_90','goals_plus_assists_minus_penalty_per_90','non_penalty_xG_plus_xA_per_90','url','league'], inplace = True)
```

Changing West Germany into Germany

```python
df['team'] = df['team'].replace('West Germany', 'Germany')
print(df['team'].unique())
```

Convert several columns to numerical columns

```python
df['season'] = pd.to_numeric(df['season'], errors='coerce')
cols = df.columns[3:24]
df[cols] = df[cols].apply(pd.to_numeric, errors ='coerce')

df.info()
```

Add a column for the winners of each season

```python
world_cup_winners = world_cup_winners = {
    1930: 'Uruguay',
    1934: 'Italy',
    1938: 'Italy',
    1950: 'Uruguay',
    1954: 'Germany',
    1958: 'Brazil',
    1962: 'Brazil',
    1966: 'England',
    1970: 'Brazil',
    1974: 'Germany',
    1978: 'Argentina',
    1982: 'Italy',
    1986: 'Argentina',
    1990: 'Germany',
    1994: 'Brazil',
    1998: 'France',
    2002: 'Brazil',
    2006: 'Italy',
    2010: 'Spain',
    2014: 'Germany',
    2018: 'France',
    2022: 'Argentina'
}


df['world_cup_winner'] = df.apply(lambda row : 1 if row['team'] == world_cup_winners.get(row['season']) else 0, axis = 1)
```

Add a continent column for each team

```python
countries = pd.read_csv('data/countryContinent.csv', encoding = 'latin1')
countries = countries[['country', 'continent']]
```

Changing some of the country names

```python
country_rep = {
    'Bolivia (Plurinational State of)':'Bolivia',
    'United States of America':'United States',
    'Macedonia (the former Yugoslav Republic of)':'Yugoslavia',
    'Czech Republic':'Czechoslovakia',
    'Indonesia':'Dutch East Indies',
    'Turkey':'Türkiye',
    'United Kingdom of Great Britain and Northern Ireland' : 'England',
    'Korea (Republic of)': 'Korea Republic',
    "Korea (Democratic People's Republic of)" : 'Korea DPR', 
    'Iran (Islamic Republic of)' : 'IR Iran',
    'Ireland' : 'Rep. of Ireland',
    'United Arab Emirates' : 'UAE',
    'Russian Federation' : 'Russia',
    'China' : 'China PR',
    'Bosnia and Herzegovina' : "Bosnia & Herz'na"
}

countries['country'] = countries['country'].replace(country_rep)
```

Merge to the dataframe

```python
df = df.merge(countries, left_on = 'team', right_on = 'country').drop(columns = 'country')
```

Dealing with missing values

```python
df.isnull().sum()
```

Threshold for missing values: drop columns with more than 50% missing values

```python
threshold = 0.50
df_cleaned = df.dropna(thresh=int(df.shape[0] * (1 - threshold)), axis=1)
```

Checking if all the missing values have been dealt with

```python
df_cleaned.isnull().sum()
```
